```python
import numpy as np
import matplotlib.pyplot as plt

ahorros =np.random.randint(50, 100, size=[6])
meses = ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio"]
mapeado = range(len(meses))


plt.plot(ahorros)
plt.xticks(mapeado, meses)
plt.show()
```

```python
plt.plot(ahorros)
plt.xticks(mapeado, meses)
plt.ylim(0,100)
plt.show()
```

```python
plt.plot(ahorros)
plt.xticks(mapeado, meses)
plt.xlim(1, 4)
plt.ylim(0,100)
plt.show()
```
